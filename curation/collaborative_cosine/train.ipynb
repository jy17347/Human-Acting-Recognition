```python
import numpy as np
import pandas as pd
import os, time
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
```

```python
height = 180
weight = 85
sex = 1
age = 20
bmi = weight*10000 /height**2 
ability = 4
height = (height-170)/30
weight = (weight-70)/30
age = (age-30)/30
bmi = (bmi-30)/20
input_data = [[height, weight, sex, age, bmi, ability]]
input_data
```

```python
dataset_dir = 'C://project//dataset/kwix/curation'
user_data = pd.read_csv(dataset_dir + '/user.csv')
user_data = user_data.to_numpy()
user_data[:,1] = (user_data[:,1]-170)/30
user_data[:,2] = (user_data[:,2]-70)/30
user_data[:,4] = (user_data[:,4]-30)/30
user_data[:,5] = (user_data[:,5]-30)/20
user_similarity = cosine_similarity(input_data,user_data[:,1:])
user_similarity
```

```python
chest_df = pd.read_csv(dataset_dir +'/chest.csv')
chest = pd.read_csv(dataset_dir +'/chest.csv').to_numpy()[:, 1:]
lower = pd.read_csv(dataset_dir + '/lower_body.csv').to_numpy()[:, 1:]
abdom = pd.read_csv(dataset_dir + '/abdominals.csv').to_numpy()[:, 1:]
```
